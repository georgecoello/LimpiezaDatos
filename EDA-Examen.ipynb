```python
cols = ['Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status',
       'Income_Category', 'Card_Category']
```

```python
valids = []
many = []
one = []

size = card.shape[0]
umbral = 0.9
```

```python
for c in cols : 
    level = card.loc[:,cols].drop_duplicates().shape[0]
    if (level/size) > umbral: 
        many.append(c)
    elif level == 1: 
        one.append(c)
    else: 
        valids.append(c)
```

Se hará el reemplazo con Education_Level

```python
level = card.groupby(["Education_Level"], as_index = False).agg({"Customer_Age":["count"]})
```

```python
level.columns = level.columns.droplevel(1)
```

```python
level
```

```python
level["Percent"] = level["Customer_Age"]/size
```

```python
level.sort_values(["Percent"], ascending = False)
```

```python
level["Education_t"] = "Others"
```

```python
level["Education_t"] = level.apply(lambda r: r["Education_Level"] if r["Percent"] > 0.1 else r["Education_t"], axis=1)
```

```python
level.loc[:,["Education_Level", "Education_t"]]
```

```python
card = card.merge(level, on= ["Education_Level"], how = "inner")
```
